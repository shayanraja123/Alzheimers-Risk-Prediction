```python
!pip install xgboost shap lime scikit-learn matplotlib seaborn

# Import libraries
import pandas as pd
import numpy as np
import xgboost as xgb
import shap
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score, StratifiedKFold
from lime.lime_tabular import LimeTabularExplainer
import seaborn as sns
```

```python
# Load validation set
validation_df = pd.read_csv("alzheimers_disease_data/alz_validation.csv")

# Define the selected 16 features and target column
features = [
    'MMSE', 'FunctionalAssessment', 'ADL', 'MemoryComplaints', 'BehavioralProblems',
    'Forgetfulness', 'Confusion', 'Disorientation', 'DifficultyCompletingTasks',
    'PersonalityChanges', 'FamilyHistoryAlzheimers', 'Depression', 'HeadInjury',
    'Hypertension', 'CardiovascularDisease', 'Diabetes'
]
target = 'Diagnosis'

# Prepare X and y
X_val = validation_df[features]
y_val = validation_df[target]
```

```python
# Define the model
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', importance_type='gain', random_state=42)

# Train the model on validation data
xgb_model.fit(X_val, y_val)
```

```python
# Plot feature importances from XGBoost
xgb.plot_importance(xgb_model, importance_type='gain', title='XGBoost Feature Importance')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Define Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Re-train model on CV splits from validation set
cv_scores = cross_val_score(xgb_model, X_val, y_val, cv=skf, scoring='accuracy')

# Print results
print("Cross-Validation Accuracy Scores:", cv_scores)
print("Mean Accuracy:", np.mean(cv_scores))
print("Standard Deviation:", np.std(cv_scores))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create containers
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
all_y_true = []
all_y_pred = []

# Loop through folds and collect predictions
for train_index, test_index in skf.split(X_val, y_val):
    X_train_fold, X_test_fold = X_val.iloc[train_index], X_val.iloc[test_index]
    y_train_fold, y_test_fold = y_val.iloc[train_index], y_val.iloc[test_index]

    # Fine-tuned XGBoost model
    model = xgb.XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42,
        max_depth=4,
        learning_rate=0.1,
        n_estimators=100,
        subsample=0.9,
        colsample_bytree=0.9,
        scale_pos_weight=1.2  # Increase sensitivity to AD class (positive class)
    )
    model.fit(X_train_fold, y_train_fold)

    # Predict probabilities and apply adjusted threshold
    y_proba_fold = model.predict_proba(X_test_fold)[:, 1]
    y_pred_fold = (y_proba_fold >= 0.45).astype(int)

    # Store results
    all_y_true.extend(y_test_fold)
    all_y_pred.extend(y_pred_fold)

# Evaluate
conf_matrix_cv = confusion_matrix(all_y_true, all_y_pred)
report_cv = classification_report(all_y_true, all_y_pred, target_names=["No AD", "AD"])
accuracy_cv = accuracy_score(all_y_true, all_y_pred)

# Print results
print("Cross-Validated Accuracy:", accuracy_cv)
print("\nClassification Report:\n", report_cv)
print("Confusion Matrix:\n", conf_matrix_cv)

# Plot confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(conf_matrix_cv, annot=True, fmt='d', cmap='Blues', xticklabels=["No AD", "AD"], yticklabels=["No AD", "AD"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Fine-Tuned XGBoost (Threshold 0.45)")
plt.show()
```

## FAIRNESS ANALYSIS: Groupwise Performance (Gender, Age, Ethnicity)

```python
# Gender

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

gender_series = validation_df.loc[:, 'Gender']


all_y_true = np.array(all_y_true)
all_y_pred = np.array(all_y_pred)


male_mask = (gender_series == 0)
male_true = all_y_true[male_mask]
male_pred = all_y_pred[male_mask]

female_mask = (gender_series == 1)
female_true = all_y_true[female_mask]
female_pred = all_y_pred[female_mask]


male_accuracy = accuracy_score(male_true, male_pred)
male_precision = precision_score(male_true, male_pred)
male_recall = recall_score(male_true, male_pred)
male_f1 = f1_score(male_true, male_pred)

female_accuracy = accuracy_score(female_true, female_pred)
female_precision = precision_score(female_true, female_pred)
female_recall = recall_score(female_true, female_pred)
female_f1 = f1_score(female_true, female_pred)



fairness_gender_df = pd.DataFrame({
    'Group': ['Male', 'Female'],
    'Accuracy': [male_accuracy, female_accuracy],
    'Precision': [male_precision, female_precision],
    'Recall': [male_recall, female_recall],
    'F1-Score': [male_f1, female_f1]
})

print("\n=== Gender Fairness Analysis ===")

display(fairness_gender_df)
```

**Analysis:**

Our fairness analysis across gender groups reveals that the XGBoost model performs slightly better for female patients compared to male patients. Specifically, for females, the model achieved an accuracy of 94.20%, precision of 91.86%, recall of 94.05%, and an F1-score of 92.94%. In contrast, for males, the model recorded an accuracy of 89.69%, precision of 89.66%, recall of 84.78%, and an F1-score of 87.15%.

These results suggest that the model is more effective at correctly identifying Alzheimer’s Disease (AD) cases among females, as indicated by the higher recall and F1-scores. While the model still performs reasonably well for males, the slightly lower recall for males indicates a somewhat higher risk of false negatives in that subgroup. Overall, there is a small but notable gender disparity, with females benefiting from slightly more reliable predictions than males.

```python
# Age

# defining age groups
def categorize_age(age):
    if age < 70:
        return "60-69"
    elif age < 80:
        return "70-79"
    elif age < 90:
        return "80-89"
    else:
        return "90+"


validation_df['AgeGroup'] = validation_df['Age'].apply(categorize_age)


age_metrics = []

for group in validation_df['AgeGroup'].unique():
    idx = validation_df['AgeGroup'] == group
    y_true_group = y_val[idx]
    y_pred_group = np.array(all_y_pred)[idx]

    acc = accuracy_score(y_true_group, y_pred_group)
    prec = precision_score(y_true_group, y_pred_group)
    rec = recall_score(y_true_group, y_pred_group)
    f1 = f1_score(y_true_group, y_pred_group)

    age_metrics.append([group, acc, prec, rec, f1])


age_metrics_df = pd.DataFrame(age_metrics, columns=["Group", "Accuracy", "Precision", "Recall", "F1-Score"])
print("\n=== Age Group Fairness Analysis ===")
display(age_metrics_df)
```

**Analysis:**

The Age Group Fairness Analysis revealed that the model performs poorly across all age ranges, with notably low precision, recall, and F1-scores. The 60-69 group achieved the highest overall accuracy (57.03%), followed closely by the 80-89 group (55.22%). However, both precision and recall remained low across groups, indicating frequent misclassifications. The 70-79 and 90+ groups had the weakest performance, with F1-scores of just 0.32 and 0.35, respectively. These results suggest that the model struggles more with age-based generalization compared to gender, and there may be underlying age-related biases or insufficient feature separability among different age groups that warrant further investigation.

```python
# Ethnicity

ethnicity_groups = validation_df['Ethnicity']

ethnicity_mapping = {
    0: 'Caucasian',
    1: 'African American',
    2: 'Asian',
    3: 'Other'
}


ethnicity_results = []

for code, label in ethnicity_mapping.items():
    idx = ethnicity_groups == code
    y_true_group = np.array(y_val[idx])
    y_pred_group = np.array(all_y_pred)[idx]
    
    if len(y_true_group) > 0:
        acc = accuracy_score(y_true_group, y_pred_group)
        report = classification_report(y_true_group, y_pred_group, target_names=["No AD", "AD"], output_dict=True)
        ethnicity_results.append({
            'Group': label,
            'Accuracy': acc,
            'Precision': report['AD']['precision'],
            'Recall': report['AD']['recall'],
            'F1-Score': report['AD']['f1-score']
        })


ethnicity_fairness_df = pd.DataFrame(ethnicity_results)
print("\n=== Ethnicity Fairness Analysis ===")
display(ethnicity_fairness_df)
```

**Analysis:**

The ethnicity-based fairness analysis shows noticeable differences in model performance across groups. The model achieved the highest performance for African American patients, with an accuracy of approximately 59.6%, a precision of 52.8%, a recall of 47.5%, and an F1-score of 0.50. Asian patients showed moderate performance, with an accuracy of about 48.7% and slightly lower recall and F1-scores compared to African Americans. In contrast, the model performed relatively poorly for Caucasian patients, achieving only 48% accuracy and a low F1-score of 0.37. The performance was worst for patients categorized as “Other” ethnicities, with just 51% accuracy, 12.5% precision, and an extremely low F1-score of 0.16, suggesting significant misclassification issues. Overall, these results indicate fairness gaps in the model, as it tends to predict Alzheimer’s outcomes more accurately for African American and Asian patients compared to Caucasian and Other groups, highlighting a potential area for further improvement.
